```python
import numpy as np
import pandas as pd
```

```python
df_data = pd.read_csv('./ds263_tx_All_Data_1805_2015_0803_191837.CSV')
```

```python
df_data
```

```python
# 先将需要的数据提取出来
df_temp = pd.DataFrame(df_data.loc[:, ['Anon Student Id', 'Student Response Subtype', 'Duration (sec)', 'Outcome']])
df_temp
```

```python
# 01.26 直接提取所有的行为数据。
df_all_temp = df_temp
# 去掉Student Response Subtype项为空的行
df_all_temp = df_all_temp.dropna(axis = 0, subset = ['Student Response Subtype'])
df_all_temp.to_csv('初步处理data_全部.csv', encoding = 'utf8')
df_all_temp
```

```python
# 01.26日，删除Choose-Sought,补充Vector, Solve行(这两个已经解决)
df_0126_temp = df_temp[df_temp['Student Response Subtype'].isin(['Axes', 'Vector', 'Variable', 'Equation', 'Choose-Principle', 'Answer', 'Whats-Wrong', 'Solve', 'Explain-Further', 'NextStepHelp'])]
# 去掉Student Response Subtype项为空的行
df_0126_temp = df_0126_temp.dropna(axis = 0, subset = ['Student Response Subtype'])
df_0126_temp.to_csv('初步处理data_0126.csv', encoding = 'utf8')
df_0126_temp
```

```python
# 01.23日，修改dropna为指定行，解决学生反应子类型“Solve”丢失的问题
df_0123_temp = df_temp[df_temp['Student Response Subtype'].isin(['Axes', 'Vector', 'Variable', 'Equation', 'Choose-Principle', 'Answer', 'Whats-Wrong', 'Solve', 'Explain-Further', 'NextStepHelp', 'Choose-Sought'])]
# 去掉Student Response Subtype项为空的行
df_0123_temp = df_0123_temp.dropna(axis = 0, subset = ['Student Response Subtype'])
df_0123_temp.to_csv('初步处理data_0123.csv', encoding = 'utf8')
df_0123_temp
```
