```python
import pandas as pd
import numpy as np
import  matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.cluster import AgglomerativeClustering
```

```python
df_data = pd.read_csv('学生频次数据_归一化_全部行为数据.csv')
df_data = df_data.iloc[:, [x for x in range(1, int(df_data.shape[1]))]]
df_data
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
%matplotlib inline
```

```python
temp_list = list(range(1, int(df_data.shape[1])))
# 外层循环逐步增大喂给特征数目
for i in range(2, 17): # 2-16个特征
    # 内层循环将不同组合的特征进行绘图
    # combination_res_list为组合后的所有结果的列表
    combination_res_list = list(itertools.combinations(temp_list,i))
    for j in range(len(combination_res_list)):
        df_data_matrix = np.array(df_data.iloc[:, list(combination_res_list[j]) ])

        # print(df_data_matrix.shape)
        # print(df_data_matrix)

        #绘制树状图m
        print("加入的特征下标是：{}".format(str(list(itertools.combinations(temp_list,i))[j])))
        plt.figure(figsize=(20,6))
        Z = linkage(df_data_matrix, method='ward', metric='euclidean')
        p = dendrogram(Z, 0)
        plt.show()
        print('----------------------------------------------------------------------------')
```
